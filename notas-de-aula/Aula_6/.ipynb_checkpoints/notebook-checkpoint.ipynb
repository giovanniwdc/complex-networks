# **Newtwork Models**

## **Erdos-Renyi (Random Graph)**

### Let us start with the Erdos-Renyi graph. This graph is composed by *N* nodes and each connection between pairs of nodes are stablished according to probability p. This is a typical case of Bernoulli process.

```python
N = 100

av_degree = 8 
p = av_degree/(N - 1)

GER = nx.gnp_random_graph(N, p, seed = 42, directed = False)
```

### Notice that the average degree of the generated network is not exactly as the input average degree. Thei input degree is the expected number of connections

```python
N = len(GER)
M = GER.number_of_edges()

print(f'Number of nodes: {N}')
print(f'Number of edges: {M}')

print(f'Average degree: {2*M/N}')
```

```python
positions = nx.fruchterman_reingold_layout(GER)
```

```python
plt.axis("off")

nx.draw_networkx_edges(GER, positions, alpha  = 0.5)
nx.draw_networkx_labels(GER, positions, font_color ='white')
nx.draw_networkx_nodes(GER, positions, node_size = 300, node_color = 'c')
```

### The degree distribution

```python
ks, Pk = degree_distribution(GER)
```

```python
Pk
```

```python
plt.plot(ks, Pk, 'bo', label = 'Data')

plt.xlabel('k')
plt.ylabel('P(k)')

plt.title('Degree Distribution')

plt.grid(True)

pk = binom.pmf(ks, N, p)

plt.plot(ks, pk, 'r', label = 'Binomial Distribution')

plt.legend()
```

### The second momment of the degree distribution

```python
k2 = momment_of_degree_distribution(GER, 2)

print(f'The second mommento of the degree distribution: {k2}')
```

### The Shannon Entropy of P(k)

```python
H = shannon_entropy(GER)

print(f'The Shannon entropy of the degree distribution: {H}')
```

### The clustering coefficient of a random network is equals to p.

```python
avc = nx.average_clustering(GER) 

print(f'Average clustering: {avc:.4f}, |p = {p}')
```

### The transitivity of G.

```python
cc = (nx.transitivity(GER))

print(f'Transitivity = {cc:.2f}')
```

```python
if nx.is_connected(GER) == True:
    
    l = nx.average_shortest_path_length(GER)
    
    print(f'Average shortest path length: {l}')

print(f'The predicted value is log(N)/log(average degree: {np.log(N)/np.log(av_degree)}')
```

### The network diameter:

```python
d = nx.diameter(GER)

print(f'Network diameter: {d}')
```

### Distribution of the average shortest path:

```python
if nx.is_connected(GER) == True:
    
    D = np.zeros((N, N))
    
    vl = []
    
    for i in np.arange(N):
        for j in np.arange(i+1, N):
            
            if i != j:
                
                aux = nx.shortest_path(GER, i, j)
                
                dij = len(aux) - 1
                
                D[i,j] = dij
                D[j, i] = dij
                
                vl.append(dij)
    x = np.arange(d+1)
    
    plt.hist(vl, bins = x, density = True)
    
    plt.title('Distribution of the gheodesic distances')
    
    plt.xlabel('Shortest path length (l)')
    plt.ylabel('P(l)')
    
    plt.grid(True)
    
else: 
    
    print('The graph has more than one connected component')
```

### The assortativity of random networks is close to zero.

```python
r = nx.degree_assortativity_coefficient(GER)

print(f'Assortativity: {r:.4f}')
```

### According to the probability of transition p, we habe the random graph present a phase transition, when the giant component emerge. In this regime, the size of the largest component increases with p. Let us show this behavior.

```python
N = 100

S = []
vk = []

seed = 0 

for c in np.arange(0, 5, 0.05):
    
    p = c/(N - 1)
    
    G = nx.fast_gnp_random_graph(N, p, seed = seed, directed = False)
    Gcc = sorted(nx.connected_components(G), key = len, reverse = True)
    
    G = G.subgraph(Gcc[0])
    
    S.append(len(G)/N)
    
    vk.append(c)
    
    seed += 1
```

```python
plt.plot(vk, S, label = 'Largest component size')

plt.xlabel('k')
plt.ylabel('S')

plt.title('Phase transition in random graphs')

plt.axvline(x = 1, color = 'r')

plt.grid(True)
```

## **Small World Networks**

```python
N = 50

av_degree = 8

k = int(av_degree/2)
p = 1

GWS = nx.watts_strogatz_graph(N, k, p, seed = 42)
```

```python
positions = nx.circular_layout(GWS)

plt.axis('off')

nx.draw_networkx_edges(GWS, positions, alpha = 0.5)
nx.draw_networkx_labels(GWS, positions, font_color = 'white')
nx.draw_networkx_nodes(GWS, positions, node_size = 300, node_color = 'black')
```

### The distance and the clustering coefficient vary according to the value of p.

```python
N = 50 

av_degree = 8

k = int(av_degree/2)

vcc = []
vl = []
vq = []

seed = 0
```

```python
for q in np.arange(0, 1, 0.02):
    
    G = nx.watts_strogatz_graph(N, k, q, seed = seed)
    
    cc = nx.average_clustering(G)
    
    vcc.append(cc)
    
    l = nx.average_shortest_path_length(G)
    
    vl.append(l)
    vq.append(q)
```

```python
plt.plot(vq, np.array(vcc)/vcc[0], label = 'Clustering' )
plt.plot(vq, np.array(vl)/vl[0], 'r', label = 'Shortest path length')

plt.xlabel('q')
plt.ylabel('$cc/cc_0$, $\ell/\ell_0/$')

plt.title('Small-World networks')

plt.grid(True)

plt.legend()
```

## **Functions**

```python
def shannon_entropy(G):
    
    k, Pk = degree_distribution(G)
    
    H = 0
    
    for p in Pk:
        
        if p > 0:
            
            H -= p*np.log2(p)
    
    return H
```

```python
def momment_of_degree_distribution(G, m):
    
    k, Pk = degree_distribution(G)
    
    M = sum((k**m)*Pk)
    
    return M
```

```python
def degree_distribution(G):
    
    vk = np.array(list(dict(G.degree()).values()))
    
    maxk = vk.max()
    minK = vk.min()
    
    kvalues = np.arange(maxk + 1)
    Pk = np.zeros(maxk + 1)
    
    for k in vk:
        
        Pk[k] += 1
        
    Pk = Pk/Pk.sum()
    
    return kvalues, Pk
```

```python
def clean_graph(G, remove_edges = True):
    
    G = G.to_undirected()
    nodes_original = G.nodes
    
    if remove_edges:
        
        G.remove_edges_from(nx.selfloop_edges(G))
    
    G_cc = sorted(nx.connected_components(G), key = len, reverse =True)
    G = G.subgraph(G_cc[0])
    G = nx.convert_node_labels_to_integers(G, first_label = 0 )
    
    return G, nodes_original
```
