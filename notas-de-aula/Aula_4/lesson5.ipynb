## **Functions**

```python
def clean_graph(G, remove_edges = True):
    
    G = G.to_undirected()
    
    if remove_edges:
        
        G.remove_edges_from(nx.selfloop_edges(G))
    
    G_cc = sorted(nx.connected_components(G), key = len, reverse =True)
    G = G.subgraph(G_cc[0])
    G = nx.convert_node_labels_to_integers(G, first_label = 0 )
    
    return G
```

```python
def accessibility(G):

    N = len(G.nodes)
    
    ks = np.array(list(dict(G.degree).values()))
    
    A = nx.adjacency_matrix(G)
    P = np.zeros((N, N), dtype = 'float')
    
    for i in np.arange(N):
        for j in np.arange(N):
            
            if ks[i] > 0:
                
                P[i, j] = A[i, j] / ks[i]
                
    P2 = expm(P) / np.exp(1)
    
    accessibility_values = np.zeros(N, dtype = float)
    
    for i in np.arange(N):
        
        accessibility = 0
        
        for j in np.arange(N):
            
            if P2[i, j] > 0:
                
                accessibility = accessibility + P2[i, j] * np.log(P2[i, j])
            
        accessibility = np.exp(- accessibility)
        
        accessibility_values[i] = accessibility
        
    return accessibility_values
```
