## **Questão 1)** A média da medida eigenvector centrality da rede USairport500 é igual a quanto? Considere apenas o maior componente.

```python
usairport = clean_graph(nx.read_edgelist('./data/usairport.txt', nodetype = int))
```

```python
usairport_EC = np.array(list(dict(nx.eigenvector_centrality(usairport, max_iter = 1000)).values()))
usairport_EC_average = usairport_EC.mean()

print(f'Hamsterster average eigenvector centrality: {usairport_EC_average:.2f}')
```

## **Questão 2)** Calcule a correlação de Pearson entre a medida betweeness centrality e grau para a rede hamsterster.

```python
hamsterster = clean_graph(nx.read_edgelist('./data/hamsterster.txt', nodetype = int))
```

```python
hamsterster_betweeness_centrality = np.array(list(dict(nx.betweenness_centrality(hamsterster)).values()))
hamstarstar_deegre_values = np.array(list(dict(hamsterster.degree).values()))
```

```python
fig = plt.figure(figsize = (8, 8))

plt.title('Hamsterster', fontsize = 20)

plt.xlabel('K')
plt.ylabel('Betweeness Centrality')

plt.plot(hamstarstar_deegre_values, hamsterster_betweeness_centrality, 'o')
```

```python
hamsterster_bck_pearson = np.corrcoef(hamstarstar_deegre_values, hamsterster_betweeness_centrality)[0, 1]

print(f'The Pearson correlation of the degree and the betweenness centrality for the Hamsterster network: {hamsterster_bck_pearson:.2f}')
```

## **Questão 3)** Calcule o coeficiente de correlação de Spearman entre as medidas closeness centrality e k-core para a rede "Jazz".

```python
jazz = clean_graph(nx.read_edgelist('./data/jazz.txt', nodetype = int))
```

```python
jazz_clc = np.array(list(dict(nx.closeness_centrality(jazz)).values()))
jazz_kc = np.array(list(dict(nx.core_number(jazz)).values()))
```

```python
fig = plt.figure(figsize = (8, 8))

plt.title('Jazz', fontsize = 20)

plt.xlabel('Closeness Centrality')
plt.ylabel('K-Core')

plt.plot(jazz_clc, jazz_kc, 'o')
```

```python
jazz_clckc_spearmanr = spearmanr(jazz_clc, jazz_kc)

print(f'The Spearmanr correlation of the Closeness Centrality and the K-Core centrality for the Hamsterster network: {jazz_clckc_spearmanr}')
```

## **Functions**

```python
def clean_graph(G, remove_edges = True):
    
    G = G.to_undirected()
    
    if remove_edges:
        
        G.remove_edges_from(nx.selfloop_edges(G))
    
    G_cc = sorted(nx.connected_components(G), key = len, reverse =True)
    G = G.subgraph(G_cc[0])
    G = nx.convert_node_labels_to_integers(G, first_label = 0 )
    
    return G
```
